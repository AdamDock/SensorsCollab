```python
import sympy as sp

r = 0.287 #kPa.m3/kg·K
FRvolume = 0.6 # m^3
DoorOpensPerDay = 20
DoorOpensPerYear = DoorOpensPerDay * 365
electricityCost = 0.140 #$/Kwh
kitchenTemp = 293.15 # K
kitchenPresssure = 95 #kPA
fridgeTemp = 277.15 #k
moistureKitchenAir = 0.010 #kg/kgair
moistureFridgeAir = 0.004 #kg/kgair
specificHeatAir = 1.005 #kJ/kg*C
fridgeCOP = 1.4
doorOpenVolumeReplacement = 0.3 #m^3
fridgeTempDoorOpenVolume = 0.3 # m^3
#TotalPrice = price per KWh * Work in kwh
# Work in Kwh = TotalHeatIntoSystem per hour / fridgeCOP -- Total heat for year divide by 365 * 24
#TotalHeatIntoSystem = Qlatent + Qsensible - Qlatent is energy for water vaper- Q sensible is Energy to raise temp using Cp

def idealGasSolver(p=None, v=None, m=None, r=None, t=None):
    P, V, M, R, T = sp.symbols('P V M R T')
    equation = sp.Eq(P * V, M * R * T)

    inputs = {P: p, V: v, M: m, R: r, T: t}

    # Filter out the 'None' value (the one we are solving for)
    known_values = {sym: val for sym, val in inputs.items() if val is not None}

    if len(known_values) != 4:
        return "Error: You must provide exactly four variables to solve for the fifth."

    missing_symbol = [sym for sym in inputs if inputs[sym] is None][0]

    solution = sp.solve(equation, missing_symbol)[0]
    result = solution.subs(known_values)

    return float(result)
hfg = 2492 #kj/kg
replacement_fridge_mass = idealGasSolver(p=kitchenPresssure, v=doorOpenVolumeReplacement, r=r, t=fridgeTemp)
latentHeat = replacement_fridge_mass * (moistureKitchenAir - moistureFridgeAir) * hfg
sensibleHeat = replacement_fridge_mass  * specificHeatAir * (kitchenTemp - fridgeTemp)
totalHeat = (latentHeat + sensibleHeat) * DoorOpensPerYear

TotalYearlyWork_kWh = (totalHeat / 3600) / fridgeCOP
TotalYearlyEnergyPrice = TotalYearlyWork_kWh * electricityCost
display(str(f'Total yearly energy price is {TotalYearlyEnergyPrice}'))
#If air is dry, not latent heat
totalHeat = sensibleHeat * DoorOpensPerYear
Work = totalHeat / fridgeCOP
TotalYearlyEnergyPriceDry = Work /3600 * electricityCost
display(str(f'Total yearly energy price if air is dry is {TotalYearlyEnergyPriceDry}'))


#The volume that 0.002 kg of liquid water takes up is only about 0.000002 m³. Negligible
```
