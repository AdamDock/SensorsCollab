```python
import numpy as np
from numpy import arange
import matplotlib.pyplot as plt
x = np.arange(1, 6)
y = np.array([1.2, 1.9, 3.2, 4.1, 5.3])
f = np.polyfit(x, y, 1, full = True)
p = f[0]
res = f[1]
yc = np.polyval(p, x)
D = np.sum((y-yc)**2)
plt.subplots(figsize = (8,3))
plt.plot(x,y,'xr')
plt.plot(x, yc, '--m')
print(D, res)
v = len(x) - (1+1)
t_4_95 = 2.776
s_yx = np.sqrt(D/v)
sb = np.sqrt(1/len(x)+(x-np.mean(x))**2/np.sum((x-np.mean(x))**2))
print(sb)
ub = yc + t_4_95 * s_yx * sb
lb = yc - t_4_95 * s_yx * sb
print(ub, lb)
plt.plot(x, ub)
plt.plot(x, lb)
plt.show()

```
