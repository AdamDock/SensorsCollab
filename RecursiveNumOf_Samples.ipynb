```python
import math
from scipy import stats
# 99% confidence for interval + or - 0.1mm
# Stdeviation is 0.08 mm
# confidence level is 99% - Significance is 1%

def initializeSamplesRequiredRecursion(nGuess, std, d, alpha):
    print(f'Standard Deviation: {std} mm')
    print(f'Confidence Interval: +-{d} mm at confidence level of {(1-alpha)*100}%')
    print(f'D value: {d}')
    calculateNWithRecursion(nGuess, std, d, alpha)

def calculateNWithRecursion(nGuess, std, d, alpha, tolerance=0.001):
    print()
    print(f'N guess is {nGuess}')
    fractional, integer = math.modf(nGuess)
    if fractional > 0:
        df = math.floor(nGuess)
    else:
        df = nGuess - 1

    df = max(1, int(df))
    print(f'Degrees of freedom {df}')

    t_val = stats.t.ppf(1 - alpha/2, df)
    print(f'T value for new calculation {t_val}')
    n_new = (t_val * std / d)**2

    print(f"Guess: {nGuess:.4f} ----> t: {t_val:.3f} --> Result: {n_new:.4f}")
    fractional, integer = math.modf(n_new)

    if abs(n_new - nGuess) < tolerance:
        print()
        print(f'Final N calculated is: {math.ceil(n_new)}')
        return n_new

    return calculateNWithRecursion(n_new, std, d, alpha, tolerance)

initializeSamplesRequiredRecursion(20, 0.08, 0.1, 0.01)



```
