## The Wheat Harvest Conundrum

For decades, farmers in the 'Golden Plains' region have meticulously recorded their wheat yields. Historically, using traditional methods, their average wheat yield has been a consistent **55 bushels per acre**.

However, since 2010, a group of innovative farmers adopted a new, sustainable farming technique. They believe this method, which focuses on soil health and biodiversity, could lead to better yields, but they want to confirm it statistically. They have collected the following average annual yields (in bushels per acre) from 2010 to 2021 using the new method:

**Yields (bushels per acre) for the new method (2010-2021):**
`[58, 60, 57, 62, 59, 61, 63, 58, 60, 64, 62, 65]`

 Use a significance level (alpha) of 0.05

```python
import numpy as np
from scipy import stats
#Null Hypothesis means it will equal the average
alpha = 0.05
mean = 55
new_sampling = np.array([90, 60, 57, 90, 80, 61, 63, 80, 60, 64, 90, 90])

# Perform a one sided T-test
def getImportantSamplingMetrics(new_sampling):
  return np.mean(new_sampling), np.std(new_sampling)
def t_test(mean, new_sampling):
  samplingMean, samplingSTD = getImportantSamplingMetrics(new_sampling)
  tSample = (samplingMean - mean)/samplingSTD * np.sqrt(len(new_sampling))
  tExpected = stats.t.ppf(1 - alpha/2, len(new_sampling) - 1)
  if(tSample > tExpected):
    print("Reject the null hypothesis")
  else:
    print("Fail to reject the null hypothesis")
t_test(mean, new_sampling)
```
