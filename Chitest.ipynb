```python
import numpy as np
from scipy.special import gamma
from numpy import exp
from scipy.integrate import quad
import matplotlib.pyplot as plt


def chi(x, v):
  return 1/2**(v/2) * 1/gamma(v/2)* x**(v/2-1) * exp(-x/2)

quad(chi, 0, 30.15, args=(11,))[0]
```
