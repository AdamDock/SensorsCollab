```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from numpy import exp, pi, sqrt

def gauss(x, xp, sig):
  return 1/sig/sqrt(2*pi)*exp(-0.5*(x-xp)**2/sig**2)
xp = 3
x = np.arange(0,5,0.01)
sig = .2
plt.plot(x,gauss(x,xp,sig))

i = quad(gauss, 2, 2.2, args = (xp, sig))
print (f'The integral is ${i}')
```
