```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.integrate import quad
from numpy import sin, cos, pi, sqrt, exp

def zfun(beta):
  return 1/sqrt(2*pi) * exp(-0.5 * beta**2)
x = np.arange(-10,10)
plt.plot(x, zfun(x))
plt.show()
quad(zfun, -1.5, -0.5)[0]
```

```python
def t_dist(x, v):
  a = gamma((v+1)/2)
  b = sqrt(v*pi)*gamma(v/2)
  c = (1+x**2/v)
  d = (-(v+1)/2)
  return a/b*c**d
quad(t_dist, 0, 2.262, args =(9,))
y = np.arange(-10, 10, .01)
plt.plot(x, zfun(x))
plt.plot(y, t_dist(y, 2))
```
